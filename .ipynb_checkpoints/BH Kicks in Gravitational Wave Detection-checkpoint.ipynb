## Briefing
___________________________________________________________________________________________________________________
This notebook is a collection of a number of simulation tests for determining the fitness of a number of calculations undertaken, such as the <b>overlap calculation</b>, and <b>the velocity shift function</b>.

The tests that are to be undertaken are: 
1. Simple FFT and IFFT of a waveform to ensure IFFT(FFT(h(t))) = h(t), also confirm the overlap calculation returns 1. The steps for this are: 
    1. import waveform, $h$
    2. $FFT(h)$
    3. $IFFT(FFT(h))$
    4. Plot the residuals $IFFT(FFT(h)) - h$
    5. calculate the overlap
2. Transform the waveform by a velocity of 0c
    1. import everything
    2. transform $h_o$ with a kick of velocity = 0c
    3. IFFT of $h_k$
    4. Plot residuals of $h_o$ and $h_k$
    5. Calculate overlap of ho and hk, should be 1
3. Transform of 0c and phase shift of $2\pi$
    1. Same steps as 2. however include a phase shift of $2\pi$
    2. Plot residuals
    3. overlap calculation
4. Set of a number of different plots and calculations

#### Imports and Initialisation

```python
%matplotlib inline

import toolsV2 as gwt
import BHKickV2 as bhk
import numpy as np
import matplotlib as plt


#Initialising the noise object
noi = bhk.noise()


#Constants used (at least initially)
sigma = 10
velocity = 0
tshift = 0
phase = 0
```

### Stage 1: $FFT \rightarrow IFFT$

```python
#Importing the waveform
wave = bhk.waveform("cbc_q1.00_M60_d410_t140.00_p0.00.dat")  

ifftwaveht = gwt.infft(wave.hf,1/wave.dT)  #Calculating the IFFT of the waveform
wavehf = gwt.fft(ifftwaveht,1/wave.dT)[1] #Calculating the re-fft of the ifft

gwt.plot(wave.t,(wave.ht-ifftwaveht),title='Residual Plot - Success') #Plotting the residual graph

#Calculation of the overlap between the FFT and the FFT of the IFFT of the FFT (because I can)
print('Overlap Calc = ' + str(gwt.overlap_calc(wavehf,wave.hf,wave.f,noi.aligochar)))
```

### Stage 2: $FFT \rightarrow v=0c \rightarrow IFFT$

```python
#Creation of the kicked waveform: 
velowave = bhk.velowaveform(wave.t,wave.ht,velocity,sigma,tshift,phase)

veloht = gwt.infft(velowave.hf,1/velowave.dT) #ifft of the velocity waveform

gwt.plot(wave.t,(wave.ht-veloht),title='Residual Plot - SUCCESS') #Plotting the residual graph
#gwt.plot(wave.t,(wave.t-velowave.stime),title='Time Difference Plot') #Plotting the time differences

gwt.wave_plot(wave.t,wave.ht,velowave.t,velowave.ht) #plotting the two waveforms

#Calculation of the overlap between the FFT and the kicked waveform
print('Overlap Calc = ' + str(gwt.overlap_calc(velowave.hf,wave.hf,wave.f,noi.aligochar)))
```

### Stage 3: Everything before + phase shift of $2\pi$

```python
#Setting of phase
phase = 2*np.pi
velocity = 0
tshift =0

#Creation of the kicked waveform: 
velowave = bhk.velowaveform(wave.t,wave.ht,velocity,sigma,tshift,phase)
veloht = gwt.infft(velowave.hf,1/velowave.dT) #ifft of the velocity waveform
gwt.plot(wave.t,(wave.ht-veloht),title='Residual Plot - Aligned') #Plotting the residual graph

#Calculation of the overlap between the FFT and the kicked waveform
print('Overlap Calc = ' + str(gwt.overlap_calc(velowave.hf,wave.hf,wave.f,noi.aligochar)))
```
